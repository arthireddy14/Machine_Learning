Decision Tree Regression

```python
import numpy as np
import matplotlib.pyplot as plt

# Units consumed
x=np.array([50,120,180,250,320,380]).reshape(-1,1)
# Corresponding electricity bill
y=np.array([500,1000,1500,1800,2500,2500])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg=DecisionTreeRegressor(max_depth=3,random_state=42)
dt_reg.fit(x,y)
x_grid=np.linspace(x.min(),x.max(),100).reshape(-1,1)
y_pred=dt_reg.predict(x)
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x,y_pred,color='red')
plt.xlabel('Units consumed')
plt.ylabel('Electricity bill')
plt.title('Decision Tree Regression-electricity bill slabs')
plt.show()
```

```python
y_test_pred=dt_reg.predict(x_test)
print(y_test.shape)
print(y_pred.shape)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, y_pred)
print("R² Score:", r2)
```

R² measures how well a regression model explains variance; R² = 1 indicates a perfect fit.For very small datasets, it may indicate overfitting rather than good generalization
