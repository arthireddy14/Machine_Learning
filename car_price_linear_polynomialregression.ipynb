```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
df=pd.read_csv('CarPrice_Assignment.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
x=df[['enginesize']]
y=df[['price']]
```

```python
plt.scatter(x,y)
plt.title('Engine size vs car price')
plt.xlabel('enginesize')
plt.ylabel('price')
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
lin_reg=LinearRegression()
```

```python
lin_reg.fit(x_train,y_train)
```

```python
y_pred=lin_reg.predict(x_test)
```

```python
print("Linear Regression")
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
print("MSE:",mse)
print("MAE:",mae)
print("R2 Score:", r2_score(y_test, y_pred))
print("RMSE:", np.sqrt(mse))
```

```python
poly=PolynomialFeatures(degree=2)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
poly_model=LinearRegression()
poly_model.fit(x_train_poly,y_train)
```

```python
y_pred_poly=poly_model.predict(x_test_poly)
```

```python
print("\nPolynomial Regression (Degree 2)")
mse1=mean_squared_error(y_test,y_pred_poly)
mae1=mean_absolute_error(y_test,y_pred_poly)
print("MSE:",mse1)
print("MAE:",mae1)
print("R2 Score:", r2_score(y_test, y_pred_poly))
print("RMSE:", np.sqrt(mse1))
```

```python
x_plot = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)
```

```python
y_lin_plot = lin_reg.predict(x_plot)

x_plot_poly = poly.transform(x_plot)
y_poly_plot = poly_model.predict(x_plot_poly)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

# Actual data
plt.scatter(x, y, label='Actual Data')

# Linear line
plt.plot(x_plot, y_lin_plot, label='Linear Regression')

# Polynomial curve
plt.plot(x_plot, y_poly_plot, label='Polynomial Regression (Degree 2)')

plt.xlabel('Engine Size')
plt.ylabel('Car Price')
plt.title('Linear vs Polynomial Regression')
plt.legend()
plt.show()
```
