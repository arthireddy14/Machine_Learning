```python
# from google.colab import drive
# drive.mount('/content/drive')

# df = pd.read_csv('/content/drive/My Drive/spam.csv', encoding='latin-1')
import pandas as pd
from google.colab import drive

drive.mount('/content/drive', force_remount=True)

df = pd.read_csv('/content/drive/My Drive/spam.csv', encoding='latin-1')
print(df.head())

```

```python
df.columns
```

```python
df.columns=['label','text']
df['label']=df['label'].map({'ham':0,'spam':1})

print(df.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.naive_bayes import MultinomialNB,GaussianNB,BernoulliNB
```

Gaussian NB

```python
df['msg_len']=df['text'].apply(len)
df['word_count']=df['text'].apply(lambda x:len(x.split()))
x_num=df[['msg_len','word_count']]
y=df['label']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_num,y,test_size=0.2,random_state=42)
```

```python
model_gnb=GaussianNB()
model_gnb.fit(x_train,y_train)
```

```python
y_pred=model_gnb.predict(x_test)
```

```python
print("Gaussian NB Accuracy : ",accuracy_score(y_test,y_pred))
```

Multinomial NB

```python
from sklearn.feature_extraction.text import CountVectorizer
x=df['text']
y=df['label']
vectorizer=CountVectorizer(stop_words='english')
x_vec=vectorizer.fit_transform(x)
x_train1,x_test1,y_train1,y_test1=train_test_split(x_vec,y,test_size=0.2,random_state=42)
```

```python
model_gnb=MultinomialNB()
model_gnb.fit(x_train1,y_train1)
```

```python
y_pred1=model_gnb.predict(x_test1)
```

```python
print("Multinomial NB Accuracy : ",accuracy_score(y_test,y_pred1))
```

Bernoulli NB

```python
vectorizer_bin=CountVectorizer(stop_words='english',binary=True)
x_bin=vectorizer_bin.fit_transform(x)
```

```python
x_train2,x_test2,y_train2,y_test2=train_test_split(x_bin,y,test_size=0.2,random_state=42)
```

```python
model_bnb=BernoulliNB()
model_bnb.fit(x_train2,y_train2)
```

```python
y_pred2=model_bnb.predict(x_test2)
print("Bernoulli NB Accuracy : ",accuracy_score(y_test2,y_pred2))
```
